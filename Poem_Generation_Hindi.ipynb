```python
import numpy as np
import pandas as pd
import os
from pathlib import Path
path = '/content/poem_dataset'; path = Path(path) #Setting the path to dataset

#Converting the dataset into Dataframe
poems = []
for file in os.listdir(path):
    filename = path/file
    f = open(filename, 'r')
    text = f.read()
    poems.append(text)

poems_df = pd.DataFrame(poems)
poems_df.head()
```

```python
!pip install fastai
```

```python
from fastai.text.all import *

# Defining a DataBlock for text data
dls_lm = DataBlock(
    blocks=TextBlock.from_df(0, is_lm=True), # Specify column 0 for text data
    get_x=ColReader(0),                      # Read data from column 0
    splitter=RandomSplitter(0.2)             # Split data randomly with 20% for validation
).dataloaders(poems_df, bs=128)
```

```python
dls_lm.show_batch(max_n=5)
```

```python
#Loading the AWD_LSTM language model
learn = language_model_learner(dls_lm, AWD_LSTM, metrics=[accuracy,Perplexity()], drop_mult=0.5)
```

```python
#Finding the best learning rate to use
learn.lr_find()
```

```python
#Plot between learning rate and loss
learn.recorder.plot_lr_find()
```

```python
#Finetuning the model for 3 epochs
learn.fit_one_cycle(3, 1e-01, moms=(0.8,0.7,0.8))
```

Prediction

```python
# 1.
predict = learn.predict('बादल', n_words = 50)
```
